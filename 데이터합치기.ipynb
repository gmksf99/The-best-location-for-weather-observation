```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

df1 = pd.read_csv('집계구별_변수_개수.csv', encoding='euc-kr')
df2 = pd.read_csv('구로_집계_공원_유수지_하천.csv', encoding='euc-kr')

df1 = df1[['TOT_REG_CD', 'area', 'density', 'value', 'workman', 'marketNum', 'industrial', 'restaurant', 'distan']]
df1
```

```python
df2 = df2[['TOT_REG_CD', 'riverNum', 'parkNum', 'waterNum']]
df2
```

```python
df_join = pd.merge(df1, df2, on="TOT_REG_CD", how="left")
df_join
```

```python
df_join.to_csv('join_data.csv', encoding='euc-kr')
```
